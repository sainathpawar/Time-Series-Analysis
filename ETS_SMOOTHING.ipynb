```python
air_df = pd.read_csv('airline_passengers.csv',index_col='Month',parse_dates=True) 
#Convert Month Column as Index Col with datetime conversion
```

```python
air_df.head()
```

```python
air_df.shape #number of rows * number of column
```

```python
#drop missing value if any, this is important before performing ETS smoothing.

air_df = air_df.dropna()
```

```python
air_df.shape
```

It means data does not contain any missing value.

```python
air_df.plot() #plot the graph
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(air_df['Thousands of Passengers'],model='multiplicative')

#data is in increasing expoential order so we will choose multiplicative model
#if series in linear in nature then we can choose additive model
```

```python
result.plot() #get two same graph for all component plot, just use ; after the code to get one set of plot
```

```python
result.plot();
```

```python
#Lets increase the size of the plot with the helps of rcParams

from pylab import rcParams
rcParams['figure.figsize'] = 12,5
result.plot();
```

```python
result.trend.plot() #you can individually plot graph for trend,seasanlity & residual
#result.seasonal.plot()
#result.resid.plot()
```

```python
result.resid.plot(figsize=(12,5)) #residual means the data which is not capture in trend & seasonality component
```
